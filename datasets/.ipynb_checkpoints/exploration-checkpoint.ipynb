```python
def resampling(dataframe, column, period):
    dataframe[column] = pd.to_datetime(dataframe[column]).dt.to_period(period)
    dataframe.set_index(column).resample(period).interpolate()
    dataframe[column] = dataframe[column].astype(str)
    dataframe[column] = pd.to_datetime(dataframe[column])
    return dataframe
```

```python
df
```

```python
resampling(df,'date','D')
```

```python
drug_d
```

```python
dataframes = [population, alc_d ,alc_s ,drug_d, hom, unemp]
```

```python
from functools import reduce
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['date'],
                                            how='outer'), dataframes)
#https://stackoverflow.com/questions/44327999/python-pandas-merge-multiple-dataframes
```

```python
df_merged
```

```python
sns.heatmap(df_merged.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_merged = df_merged.dropna()
```

```python
sns.heatmap(df_merged.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_merged.head()
```

```python
df_merged.corr()
```

```python
# we want to change everything to be x/population
```

```python
df_merged[['alcohol_deaths','alcohol_sales','drug_deaths','homicides','perc_unemp']] = df_merged[['alcohol_deaths','alcohol_sales','drug_deaths','homicides','perc_unemp']].div(df_merged.us_pop, axis=0)
#https://stackoverflow.com/questions/34540567/divide-multiple-columns-by-another-column-in-pandas
```

```python
df_merged.head()
```

```python
df_merged.corr()
```

```python
sns.lineplot( x = "date", y = "alcohol_deaths", data = df_merged,markers = True)
```

```python
for x in df_merged:
    sns.lineplot( x = "date", y = x, data = df_merged,markers = True)
    plt.show()
```

```python
x = df_merged.date
plt.plot(x , df_merged.alcohol_deaths, label = "alcohol deaths")
plt.plot(x, df_merged.alcohol_sales, label = "alcohol sales")
plt.plot(x, df_merged.drug_deaths, label = "drug deaths")
#plt.plot(x, test.us_pop, label = " 4")
plt.plot(x, df_merged.perc_unemp, label = "unemployment")
plt.plot(x, df_merged.homicides, label = "homicides")
plt.legend()
plt.show()       
```

```python
#function for the same chary
for col in df_merged.columns:
    if not col == 'date' and not col == 'us_pop':
        plt.plot(df_merged['date'], df_merged[col], label=col)
plt.legend()
plt.show()
```

The above plots show the change of data over time in respect to US population; however, the differing scales make it hard to plot all of the data on the same chart. To solve this we can normalize the data. Normalizing the data will aslo prepare us run more machine learning methods on the data that might increase our knowledge of the data.
